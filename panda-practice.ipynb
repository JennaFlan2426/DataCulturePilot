Question #3: Use describe() to show some summary statistics about GDP per cap for countries in the Americas

Bonus: display each year as a row and each summary statistic as a column

```python
mask = df['continent'] == 'Americas'
df_amer = df[mask]
df_amer.loc[:,'gdpPercap_1952':'gdpPercap_2007'].describe().T
```

Question #5: First, find the minimum life expectancies for all countries with a higher than average GDP (for that year). Then, find the maximum life expectancies for all countries with a lower than average GDP (for that year)

```python
df_gdp = df.loc[:,'gdpPercap_1952':'gdpPercap_2007']
mask = df_gdp > df_gdp.mean()
df_gdp[mask]
```
